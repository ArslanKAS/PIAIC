# 1: Average Income vs Women Scatterplot

```python
state0 = pd.read_csv("states0.csv", usecols=['State','Income','GenderPop'],)
state0
```

```python
x_states = state0['Income'].str.slice(start=1)
genders = state0.GenderPop.str.split('_')
# state0[Men] = gender_split.str.get(0)
women = genders.str.get(1)
women = women.str[:-1]
# y_women = state0['GenderPop'].str.rsplit('_', 1)
y_women = women.astype('float64')
x_states = x_states.astype('float64')
```

```python
state0_DF = pd.DataFrame({'States Average Income ($)':x_states, 'Women in Population (Millon)': y_women})
```

```python
state0_DF.plot.scatter(x='States Average Income ($)',y='Women in Population (Millon)')
```
